### Visualizing Relationships

Let’s take a look at the distribution of body mass by species

```r
library(tidyverse)
library(palmerpenguins)
library(ggthemes)
library(ggplot2)
```

```r
ggplot(
  penguins,
  aes(x = species, y = body_mass_g)
) +
  geom_boxplot()
```

Alternatively, we can make density plots with

```r
ggplot(
  penguins,
  aes(x = body_mass_g, color = species, fill = species)
) +
  geom_density(linewidth = 0.75, alpha = 0.5)
```

We can use stacked bar plots to visualize the relationship between two categorical variables

```r
ggplot(
  penguins,
  aes(x = island, fill = species)
) +
  geom_bar()
```

Scatterplots are the most commonly used plot for visualizing the relationship between two numerical variables.

```r
ggplot(
  penguins,
  aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point()
```

We can incorporate more variables into a plot by mapping them to additional aesthetics.

```r
ggplot(
  penguins,
  aes(
    x = flipper_length_mm,
    y = body_mass_g
  )
) +
  geom_point(aes(color = species, shape = island)) + 
  facet_wrap(~island)
```
